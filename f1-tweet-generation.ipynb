```python
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset="all_tweets.csv",
              model_name='355M',
              steps=2000,
              restore_from='fresh',
              run_name='run1',
              print_every=10,
              only_train_transformer_layers = True,
              accumulate_gradients = 1,
              sample_every=500,
              save_every=500
              )
```

```python
gen_file = 'gpt2_gentext_{:%Y%m%d_%H%M%S}_2.txt'.format(datetime.utcnow())

gpt2.generate_to_file(sess,
                      destination_path=gen_file,
                      length=200,
                      temperature=1.0,
                      top_p=0.9,
                      prefix='<|startoftext|>',
                      truncate='<|endoftext|>',
                      include_prefix=False,
                      nsamples=2000,
                      batch_size=20
                      )
```
